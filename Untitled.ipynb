```python
#all dependices

import pandas as pd
import numpy as np
import re
import os
from pathlib import Path
```

```python
#import the datafile

df = pd.read_csv("/home/navneet/Documents/preprocessing/test.csv")
df.head()
```

```python
#step 6 pincode length and remove any alphabet from pincode column
df['POSTCODE'] = df['POSTCODE'].astype(str).str.extract('(\d{6})', expand=False)
df.head()
```

```python
#Step 1 - Filtering out of Keywords 
#find village and post keyword and replace it by space

for col in df:
    df[col]= df[col].replace('village', " ")
    df[col]= df[col].replace('post', " ")
df.head(20)
```

```python
#step 5 to remove special char from the database

import re
lis=[]
pattern = r'[^A-Za-z0-9 ]'
regex = re.compile(pattern)

for col in df:
    s=df[col].tolist()
    for i in s:
        lis.append(str(regex.sub('', str(i))))
r=pd.DataFrame(lis)
```
