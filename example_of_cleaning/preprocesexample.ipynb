```python
#Step 1 - Filtering out of Keywords 
#find village and post keyword and replace it by space

for col in dataf:
    dataf[col]= dataf[col].replace('village', " ")
    dataf[col]= dataf[col].replace('post', " ")
    dataf[col]= dataf[col].replace('Village', " ")
    dataf[col]= dataf[col].replace('Post', " ")
    dataf[col]= dataf[col].replace('VILLAGE', " ")
    dataf[col]= dataf[col].replace('POST', " ")
dataf.update(dataf)
dataf.head(6)
```
